```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
import os 

PREPROCESSED_DATA_DIR = os.path.join(os.getcwd(), "data/preprocessed_data/")
```

```python
images_names = os.listdir(PREPROCESSED_DATA_DIR)
print(f"Total images: {len(images_names)}")
```

```python
image_path = os.path.join(PREPROCESSED_DATA_DIR, images_names[-1])
print(image_path)
```

```python
original_image = cv2.imread(image_path)
plt.imshow(original_image, cmap="gray")

image = original_image.copy()
```

```python
words_data = pytesseract.image_to_string(image)
```

```python
def format_data(data:str): 

    words_list = data.strip().split("\n")
    # Remove empty lines 
    words_list = list(filter(lambda x: len(x) != 0, words_list))

    return words_list
```

```python
data = format_data(words_data)
len(data)
```
